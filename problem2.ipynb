```python
import numpy as np
import pandas as pd

'''from sklearn import '''
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
import glob
path = r'bbFiles' # use your path
allFiles = glob.glob(path + "/*")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, names = ["Year","Team","League","PlayerID","PlayerName","TotalPlayed","GamesStarted","GamesBatted",
                                                            "PlayerInDefenseTeam","AsPitcher","AsCatcher","AsFirstbaseman","AsSecondbaseman","AsThirdbaseman",
                                                            "AsShortstop","AsLeftfielder","AsCenterfielder","AsRightFielder","AsOutfielder","AsDesignatedHitter",
                                                            "AsPinchHitter","AsPinchRunner"],error_bad_lines=False, encoding="latin-1")
    list_.append(df)
frame = pd.concat(list_)
```

```python
# unique Team names
g = df1['Team'].unique()
s = pd.Series(g)
s = s.sort_values()
s.to_csv('uniqueTeam.csv', encoding='utf-8',index=None)
```
